```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the dataset using pandas library

```python
# Load the Iris dataset from seaborn
iris_df = sns.load_dataset('iris')
```

```python
# Display dataset
iris_df
```

## Display dataset structure

```python
# Display dataset shape
print("Dataset Shape:")
print(iris_df.shape)
```

```python
# Display column names
print("Dataset Columns:")
print(iris_df.columns)
```

```python
# Display first few rows
print("First 5 rows of the dataset:")
print(iris_df.head())
```

## Basic Data Summarization

```python
# Display basic information about the dataset
print("Dataset Info:")
print(iris_df.info())
```

```python
# Display statistical summary
print("Statistical Summary:")
print(iris_df.describe())
```

```python
# Display species distribution
print("Species Distribution:")
print(iris_df['species'].value_counts())
```

## Create Basic Visualizations

### Scatter Plot to analyze relationships between variables

```python
# Create scatter plot: sepal_length vs sepal_width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Create scatter plot: petal_length vs petal_width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='petal_length', y='petal_width', hue='species')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Pairwise scatter plot matrix
plt.figure(figsize=(12, 8))
sns.pairplot(iris_df, hue='species', diag_kind='hist')
plt.suptitle('Pairwise Scatter Plot Matrix', y=1.02)
plt.show()
```

### Histogram to examine data distribution

```python
# Create histograms for all numerical features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Sepal length histogram
axes[0,0].hist(iris_df['sepal_length'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Distribution of Sepal Length')
axes[0,0].set_xlabel('Sepal Length (cm)')
axes[0,0].set_ylabel('Frequency')

# Sepal width histogram
axes[0,1].hist(iris_df['sepal_width'], bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
axes[0,1].set_title('Distribution of Sepal Width')
axes[0,1].set_xlabel('Sepal Width (cm)')
axes[0,1].set_ylabel('Frequency')

# Petal length histogram
axes[1,0].hist(iris_df['petal_length'], bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Distribution of Petal Length')
axes[1,0].set_xlabel('Petal Length (cm)')
axes[1,0].set_ylabel('Frequency')

# Petal width histogram
axes[1,1].hist(iris_df['petal_width'], bins=15, color='gold', edgecolor='black', alpha=0.7)
axes[1,1].set_title('Distribution of Petal Width')
axes[1,1].set_xlabel('Petal Width (cm)')
axes[1,1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Histograms by species using seaborn
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.histplot(data=iris_df, x='sepal_length', hue='species', ax=axes[0,0], kde=True)
axes[0,0].set_title('Sepal Length Distribution by Species')

sns.histplot(data=iris_df, x='sepal_width', hue='species', ax=axes[0,1], kde=True)
axes[0,1].set_title('Sepal Width Distribution by Species')

sns.histplot(data=iris_df, x='petal_length', hue='species', ax=axes[1,0], kde=True)
axes[1,0].set_title('Petal Length Distribution by Species')

sns.histplot(data=iris_df, x='petal_width', hue='species', ax=axes[1,1], kde=True)
axes[1,1].set_title('Petal Width Distribution by Species')

plt.tight_layout()
plt.show()
```

### Box Plot to detect outliers and spread of values

```python
# Create box plots for all numerical features
plt.figure(figsize=(12, 8))
```

```python
# Single subplot with all features
plt.subplot(2, 2, 1)
sns.boxplot(data=iris_df, y='sepal_length')
plt.title('Box Plot: Sepal Length')

plt.subplot(2, 2, 2)
sns.boxplot(data=iris_df, y='sepal_width')
plt.title('Box Plot: Sepal Width')

plt.subplot(2, 2, 3)
sns.boxplot(data=iris_df, y='petal_length')
plt.title('Box Plot: Petal Length')

plt.subplot(2, 2, 4)
sns.boxplot(data=iris_df, y='petal_width')
plt.title('Box Plot: Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Box plots by species
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.boxplot(data=iris_df, x='species', y='sepal_length', ax=axes[0,0])
axes[0,0].set_title('Sepal Length by Species')

sns.boxplot(data=iris_df, x='species', y='sepal_width', ax=axes[0,1])
axes[0,1].set_title('Sepal Width by Species')

sns.boxplot(data=iris_df, x='species', y='petal_length', ax=axes[1,0])
axes[1,0].set_title('Petal Length by Species')

sns.boxplot(data=iris_df, x='species', y='petal_width', ax=axes[1,1])
axes[1,1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
# Combined box plot for all features
plt.figure(figsize=(12, 6))
iris_melted = iris_df.melt(id_vars=['species'], 
                          value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                          var_name='feature', value_name='value')

sns.boxplot(data=iris_melted, x='feature', y='value', hue='species')
plt.title('Box Plots of All Features by Species')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
